# Storytelling Data Visualization on Exchange Rates

This short project with Dataquest is designed to visualize Exchange Rates between EUR and other currencies between 1999 and 2020.
With this project we are going to import EUR exchange rates from the ECB (European Central Bank), take an exploratory look at the data, then decide what kind of analysis we are going to do.

## Data import and initial exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

```python
exchange_rates.describe()
```

1. This dataset has 5699 rows and 41 columns;
2. Period/Unit provides daily dates, this dataset is daily-granular;
3. All objects except for 3 columns, most columns should be float64 (like Turkish lira, Iceland krona and Romanian leu), except for Period/Unit which should be a timestamp. We will need to clean-up the rate columns as well as convert into timestamp Period/Unit;
4. There are null values in some columns like: Chinese yuan and Cypriot pound.

## Data Cleaning

As noted above we need to:
1. edit headers using snake notation and more intuitive names in general;
2. convert Period/Unit into timestamp;
3. convert US_dollar column (and eventually other currencies columns) into numeric (float) format.

```python
# renaming key columns
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
# edit time into timestamp format
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
# we need to clean-up some values in the dollar column before converting to float
euro_to_dollar['US_dollar'].value_counts() # 62 '-' characters
```

We can see a bunch of values '-' that are not numeric values relevant for our purporses. We need to clean those values up before converting into float.

```python
# Keeping only the valid values in both dataframes and convert into float
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

Now that data is cleaner we can start visualizing our data:

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

Let's take a look at rolling averages instead of single daily values:

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

We will keep the 30-day rolling average for our visualization purposes:

```python
euro_to_dollar['rolling_mean_30'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean_30'])
plt.show()
```

Let's extract years from Time column to have them handy when slicing our DataFrame:

```python
# extract year
euro_to_dollar['Year'] = pd.DatetimeIndex(euro_to_dollar['Time']).year
euro_to_dollar.head()
```

## Data Visualization

Ideas for the data viz project:

1. We show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline. We can use a line plot.
2. We show how the euro-dollar rate changed during the 2007-2008 financial crisis. We can also show the data for 2006 and 2009 for comparison. We can use a line plot.
3. We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.

We decide to go with Option 3, looking into the EUR-USD trend across the last 3 Presidents

### Presidential Mandates

We create some subsets of the original DataFrame for each president. We also get EUR-USD average value during each President's tenure:

## Next Steps

We are going to look into EUR exchange with the main European economies outside Euro area.

### EUR vs Other European Currencies

1. Norwegian krone: NOK 
2. Swiss franc: CHF
3. Swedish krona: SEK 
4. UK pound sterling: GBP

```python
# renaming key columns 
exchange_rates.rename(columns={'[Swiss franc ]': 'CHF_EUR',
                               '[Norwegian krone ]': 'NOK_EUR',
                               '[Swedish krona ]': 'SEK_EUR',
                               '[UK pound sterling ]': 'GBP_EUR',
                              },
                      inplace=True)
```

```python
euro_to_other = exchange_rates[['Time', 'CHF_EUR','GBP_EUR','SEK_EUR','NOK_EUR']].copy()
```

```python
# we need to clean-up some values in the dollar column before converting to float
euro_to_other['CHF_EUR'].value_counts()
# print(euro_to_other['GBP_EUR'].value_counts())
# print(euro_to_other['SEK_EUR'].value_counts())
# print(euro_to_other['NOK_EUR'].value_counts())
# 62 '-' characters 
# same as with USD, we might be able to clean everything with the same rule

euro_to_other = euro_to_other[(euro_to_other['CHF_EUR'] != '-')& (euro_to_other['GBP_EUR'] != '-')
                              & (euro_to_other['NOK_EUR'] != '-') & (euro_to_other['SEK_EUR'] != '-') ]
euro_to_other['CHF_EUR'] = euro_to_other['CHF_EUR'].astype(float)
euro_to_other['GBP_EUR'] = euro_to_other['GBP_EUR'].astype(float)
euro_to_other['NOK_EUR'] = euro_to_other['NOK_EUR'].astype(float)
euro_to_other['SEK_EUR'] = euro_to_other['SEK_EUR'].astype(float)
euro_to_other.info()
```

```python
euro_to_other.head()
```

We decide to keep only years from 2002 (physical EUR launch) to 2020

```python
# extract year
euro_to_other['Year'] = pd.DatetimeIndex(euro_to_other['Time']).year
euro_to_other.head()
```

```python
euro_to_other = euro_to_other.copy()[(euro_to_other['Year']>=2002) & (euro_to_other['Year']<=2020)]
```

```python
# prepare rolling means for all currencies
euro_to_other['CHF_rolling_30'] = euro_to_other['CHF_EUR'].rolling(30).mean()
euro_to_other['GBP_rolling_30'] = euro_to_other['GBP_EUR'].rolling(30).mean()
euro_to_other['NOK_rolling_30'] = euro_to_other['NOK_EUR'].rolling(30).mean()
euro_to_other['SEK_rolling_30'] = euro_to_other['SEK_EUR'].rolling(30).mean()
```

```python
euro_to_chf_avg = round(euro_to_other['CHF_rolling_30'].mean(),2)
print(euro_to_chf_avg)

euro_to_gbp_avg = round(euro_to_other['GBP_rolling_30'].mean(),2)
print(euro_to_gbp_avg)

euro_to_nok_avg = round(euro_to_other['NOK_rolling_30'].mean(),2)
print(euro_to_nok_avg)

euro_to_sek_avg = round(euro_to_other['SEK_rolling_30'].mean(),2)
print(euro_to_sek_avg)
```

```python
# CHF
plt.plot(euro_to_other['Time'],
         euro_to_other['CHF_rolling_30'])
plt.show()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=4, ncols=1,
                                        figsize=(12,20))
currency_list = ['CHF_rolling_30','GBP_rolling_30','NOK_rolling_30','SEK_rolling_30']
currencies =['CHF','GBP','NOK','SEK']
adjustments_title = [+0.05, +0, +0.05, 0.05]
euro_to_other_avg = [euro_to_chf_avg, euro_to_gbp_avg, euro_to_nok_avg, euro_to_sek_avg]
axes = [ax1, ax2, ax3, ax4]
graph_colors = ['#af0b1e','#409caf','#ebc55e','#386641']
for ax, fx, color, currency, avg, adj in zip(axes, currency_list, graph_colors, currencies,
                                        euro_to_other_avg, adjustments_title):
    ax.plot(euro_to_other['Time'], euro_to_other[fx],
            color=color,linewidth=2 )
    # ax.set_yticklabels([])
    # Add some transparency to the grid
    ax.grid(alpha=0.3)
    # add title
    ax.text(730120, max(ax.get_yticks())-adj, 
            "EUR-"+currency+" Average between 2000 and 2020: " +str(avg) ,weight='bold')
ax4.text(730120, 7.6, 'Â©DATAQUEST' + ' '*190+ 'Source: European Central Bank'
         ,color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)
plt.show()
```

We can quickly notice a couple of things:
1. CHF and GBP have very negatively correlated trends, when one currency gets stronger vs EUR the other one gets weaker
2. SEK and NOK have very similar trends, the only difference is at the peak of the Financial Crisis and Sovereing Debt the Swedish currency lost more value vs EUR compared to NOK. Could it be because the Norwegian economy is considered more stable, being backed by the sovereign fund (oil funded)?
3. Brexit seems to be a major factor in GBP fluctuation vs EUR since 2016. Another check could be looking into correlations/trends between 2000 and 2015 only.

```python
# let's check correlations then:
euro_to_other.corr()
```

```python
euro_to_other_2015 = euro_to_other.copy()[(euro_to_other['Year']>=2002) & (euro_to_other['Year']<=2015)]
```

```python
euro_to_other_2015.corr()
```

1. Our visual intuition seems confirmed: GBP and CHF have a negative 0.6 correlation, while NOK and SEK have a positive 0.75, which is pretty strong! 
2. Brexit does not seem to be playing a role in CHF vs GBP correlations, it only makes it a bit weaker
3. There seemed to be some positive correlation between GBP vs EUR and the Scandinavia currencies vs EUR, but most of it seems to be happening since 2016, limiting the dataset to 2002-2015 period the correlations gets weaker.
